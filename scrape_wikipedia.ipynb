```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = 'https://packtpub.com'
```

# Viewing Source code

```python
def show_html(url):
    body = requests.get(url)
    soup = bs(body.text, 'html.parser')
    for js in soup('script'):
        js.decompose()
    return soup.prettify()
```

# Extracting Hyperlinks from sourcecode

```python
body = requests.get(url)
soup = bs(body.text, 'html.parser')
links = soup.find_all('a')
```

```python
for link in links:
    print(link.text)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
soup = bs(requests.get(url).text, 'lxml')
```

```python
table = soup.find(id='constituents')
```

```python
print(table.text)
```

```python
file = open('sp500.txt', 'w')
```

```python
type(table)
```

```python
for row in table.find_all('tr'):
    row=row.text.split('\n')
    if len(row)==12:
        row.append("NA")
    if len(row) > 13:
        row.pop()
    for column in row[1:]:
        file.write(column+ '\t')
    file.write('\n')
```
